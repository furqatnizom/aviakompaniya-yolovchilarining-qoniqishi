```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
tr_df = pd.read_csv("train_dataset.csv")
tr_df
```

```python
test_df = pd.read_csv("test_dataset.csv")
test_df
```

```python
tr_df.columns
```

```python
cat_cols = ['Gender', 'Customer Type','Type of Travel', 'Class']
```

```python
def encode_df(dataframe):
    le = LabelEncoder()
    for column in dataframe.columns:
        if column in cat_cols:
            dataframe[column] = le.fit_transform(dataframe[column])
    return dataframe
```

```python
tr_df = encode_df(tr_df)
test_df = encode_df(test_df)
```

```python
tr_df.info()
```

```python
tr_df["Arrival Delay in Minutes"].fillna(tr_df["Arrival Delay in Minutes"].mean(),inplace=True)
test_df["Arrival Delay in Minutes"].fillna(test_df["Arrival Delay in Minutes"].mean(),inplace=True)
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(tr_df.drop("satisfaction", axis=1), y=tr_df["satisfaction"])
```

```python
pred = dtc.predict(test_df)
pred
```

```python
sub = pd.read_csv("sample_submission.csv")
sub
```

```python
sub["satisfaction"] = pred
sub
```

```python
sub.to_csv("sub-avia.csv",index=False)
```

```python
rfc = RandomForestClassifier()
rfc.fit(tr_df.drop("satisfaction", axis=1), y=tr_df["satisfaction"])
```

```python
pr_y = rfc.predict(test_df)
pr_y
```

```python
sub["satisfaction"] = pr_y
sub
```

```python
sub.to_csv("sub-avia2.csv",index=False)
```
