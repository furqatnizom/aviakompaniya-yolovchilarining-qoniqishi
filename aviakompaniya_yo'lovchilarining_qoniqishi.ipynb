```python
import pandas as pd
import numpy as np
```

```python
train_data = pd.read_csv("/content/train_dataset.csv")
train_data
```

```python
test_data = pd.read_csv("/content/test_dataset.csv")
test_data
```

```python
train_data.corrwith(train_data['satisfaction']).sort_values(ascending=False)
```

```python
train_data['Gender'].value_counts()
```

```python
train_data['Gender'].replace("Female",0, inplace=True)
train_data['Gender'].replace("Male",1, inplace=True)
```

```python
test_data['Gender'].replace("Female",0, inplace=True)
test_data['Gender'].replace("Male",1, inplace=True)
```

```python
train_data['Customer Type'].value_counts()
```

```python
train_data['Customer Type'].replace("Loyal Customer",1, inplace=True)
train_data['Customer Type'].replace("disloyal Customer",0, inplace=True)
```

```python
test_data['Customer Type'].replace("Loyal Customer",1, inplace=True)
test_data['Customer Type'].replace("disloyal Customer",0, inplace=True)
```

```python
train_data['Type of Travel'].value_counts()
```

```python
train_data['Type of Travel'].replace("Personal Travel",0, inplace=True)
train_data['Type of Travel'].replace("Business travel",1, inplace=True)
```

```python
test_data['Type of Travel'].replace("Personal Travel",0, inplace=True)
test_data['Type of Travel'].replace("Business travel",1, inplace=True)
```

```python
train_data['Class'].value_counts()
```

```python
train_data['Class'].replace("Eco",0, inplace=True)
train_data['Class'].replace("Eco Plus",1, inplace=True)
train_data['Class'].replace("Business",2, inplace=True)
```

```python
test_data['Class'].replace("Eco",0, inplace=True)
test_data['Class'].replace("Eco Plus",1, inplace=True)
test_data['Class'].replace("Business",2, inplace=True)
```

```python
train_data.corrwith(train_data['satisfaction']).sort_values(ascending=False)
```

```python
X_train = train_data[['Class', 'Online boarding', 'Type of Travel',
                      'Inflight entertainment', 'Seat comfort', 'On-board service',
                      'Leg room service', 'Cleanliness', 'Flight Distance',
                      'Inflight wifi service', 'Baggage handling', 'Inflight service',
                      'Checkin service', 'Food and drink', 'Customer Type',
                      'Ease of Online booking', 'Age']]
```

```python
X_test = test_data[['Class', 'Online boarding', 'Type of Travel',
                      'Inflight entertainment', 'Seat comfort', 'On-board service',
                      'Leg room service', 'Cleanliness', 'Flight Distance',
                      'Inflight wifi service', 'Baggage handling', 'Inflight service',
                      'Checkin service', 'Food and drink', 'Customer Type',
                      'Ease of Online booking', 'Age']]
```

```python
y_train = train_data['satisfaction']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_test_scaled = scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train_scaled, y_train)
```

```python
predicted_y = model.predict(X_test_scaled)
predicted_y
```

```python
predicted_y[predicted_y>=0.5]=1
predicted_y[predicted_y<0.5]=0
```

```python
predicted_y = predicted_y.astype(np.int16)
```

```python
predicted_y.shape
```

```python
result = pd.DataFrame()
result.index = test_data['id']
result['satisfaction'] = predicted_y
```

```python
result
```
